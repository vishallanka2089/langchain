## Importing API Key

```python
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
print(api_key)  # just to test
```

## Importing Model

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.6
)
```

## Calling model and Testing


```python
response = llm.invoke("Suggest a fancy restaurant name for Indian food")
print(response)
print(response.content)
```
